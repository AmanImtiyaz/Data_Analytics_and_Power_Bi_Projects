```python
# Data Analysis and Visualization
```

```python
Cancelled_Percentage = df1['is_canceled'].value_counts(normalize = True)
Cancelled_Percentage
```

```python
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not_cancelled','Cancelled'],df1['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.countplot(x = 'hotel', data = df1, hue='is_canceled')
plt.title('Reservation status in different hotel')
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled','Cancelled'],fontsize=8)
plt.show()
```

```python
Resort_Hotel = df1[df1['hotel'] == 'Resort Hotel']
Resort_Hotel['is_canceled'].value_counts(normalize = True)
```

```python
City_Hotel = df1[df1['hotel'] == 'City Hotel']
City_Hotel['is_canceled'].value_counts(normalize = True)
```

```python
Resort_Hotel = Resort_Hotel.groupby('reservation_status_date')[['adr']].mean()
City_Hotel = City_Hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate', fontsize = 40)
plt.plot(Resort_Hotel.index, Resort_Hotel['adr'])
plt.plot(City_Hotel.index, City_Hotel['adr'])
plt.legend(['Resort Hotel','City Hotel'])
plt.show()
```

```python
df1['month'] = df1['reservation_status_date'].dt.month
plt.figure()
sns.countplot(x = 'month', data = df1, hue = 'is_canceled')
plt.legend(['Not Cancelled','Cancelled'])
plt.ylabel('Reservation count')
plt.xlabel('Month')
plt.show()
```

```python
plt.figure()
sns.barplot('month', 'adr', data = df1[df1['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data = df1[df1['is_canceled'] == 1]
Top10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.pie(Top10_country, autopct = '%.2f', labels = Top10_country.index)
plt.show()
```
